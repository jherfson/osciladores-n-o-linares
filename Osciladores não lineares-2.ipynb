```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Oscilador pêndulo plano
theta = np.linspace(-10.0, 10.0, 100)
# m é massa
m = 1
# gravidade
g = 9.8
force = -m*g*np.sin(theta)
print(force)
```

```python
plt.figure()
plt.plot(theta, force)
plt.xlabel(r'$\theta$', fontsize=15)
plt.ylabel('F', fontsize=15)
plt.tick_params(labelsize=15)
plt.show()
```
